```python
import numpy as np
import json
import os

ids = np.array([1240610, 1210718, 1210702, 1200562, 6212124, 1200575, 1200570, 1200571, 
                1240903, 1210734, 1210707, 1200558, 1200554, 1210683, 7240919, 1210770, 
                1240566, 1240675])

superTotalImages = 0
superTotalJson = 0
ImagesDirectory = 'dataset/images/'

print("____________________JSON_______________________")

for stationJson in ids:
    currentDir = 'dataset/precipitation/'+str(stationJson)+'.json'
    with open(currentDir) as f:
        # Load the JSON data
        data = json.load(f)
    f.close()
    # Create an empty dictionary to hold the results
    hour_counts = {}
    input_dict = data
    # Loop over each key-value pair in the dictionary
    for date_dict in input_dict.values():
        for value in date_dict.values():
            # Increment the count for this date
            date_str = list(input_dict.keys())[list(input_dict.values()).index(date_dict)]
            if date_str not in hour_counts:
                hour_counts[date_str] = 0
            hour_counts[date_str] += 1
    # Print the results
    total = 0
    for date_str, count in hour_counts.items():
        total = total + count
        #print(date_str, count)       
    print(f"Station ID = {stationJson} | Number of data = {total}")
    superTotalJson = superTotalJson + total


print("____________________IMAGES_______________________")
    
for stationId in ids: #para cada estação
    Num_Images_Per_Station = 0
    for stationFilename in os.listdir(ImagesDirectory+str(stationId)): #dentro de cada estação, para cada dia
        #for SpecificImage in os.listdir(ImagesDirectory+stationId)+"/"+stationFilename): #para cada dia, para cada imagem em especifico
        num_files = len([f for f in os.listdir(ImagesDirectory+str(stationId)+"/"+stationFilename)])
        Num_Images_Per_Station = Num_Images_Per_Station + num_files
    print(f"Station ID = {stationId} | Number of data = {Num_Images_Per_Station}")
    superTotalImages = superTotalImages + Num_Images_Per_Station

print("___________________________________________")
print(f"Final count Json: {superTotalJson}")
print(f"Final count Images: {superTotalImages}")
```
