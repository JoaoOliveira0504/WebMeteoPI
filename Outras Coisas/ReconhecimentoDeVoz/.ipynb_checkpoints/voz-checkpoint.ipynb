```python
# --== Imports ==--
import pyaudio
import wave
import time
import os

# --== Configuração do pyaudio ==--

# Formato do wav
FORMAT = pyaudio.paInt16
# Canais de audio. Por exemplo mono, stereo ou surround. No caso é mono
CHANNELS = 1
# Taxa de amostragem de 44.1 kHz. Esta variavel representa a quantidade de vezes que o audio vai ser captado por segundo
RATE = 44100
#Indica o numero de frames de audio por buffer. Este valor representa o tamanho do buffer utilizado para guardar o audio que esta a ser lido
CHUNK = 1024
# Segundos a ser registado
RECORD_SECONDS = 15

# --== Gravação de audio ==--

word = "zero"
WAVE_OUTPUT_FILENAME = "audio//"+word+".wav"

#Criação duma instância do PyAudio
p = pyaudio.PyAudio()
stream = p.open(format=FORMAT,
               channels=CHANNELS,
               rate=RATE,
               input=True,
               frames_per_buffer=CHUNK)

print("Recorder>> Recording...")

frames = []

# Adiciona um frame a cada "momento" do audio
'''
este for vai de 0 a int(RATE / CHUNK * RECORD_SECONDS)
como são captados 44100 frames de audio por segundo mas o buffer só aguenta
1024 este for define quantos ciclos são precisos para captar os 15 segundos de audio
ou seja um segundo de audio vai demorar varios ciclos pra ser gravado.
o int é pq não podemos deixar um buffer a meio.
demorei bastante até conseguir perceber bem este codigo :')
'''
for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK) #guarda  no variavel data o 'bocado' de audio. Esta variavel é reescrita a cada ciclo
    frames.append(data) #adiciona a variavel anterior no ficheiro final
    time_left = RECORD_SECONDS - (i * CHUNK) / RATE # calcula os segundos que faltam. Codigo do Edgar
    print(f"\rRecorder>> Time left: {int(time_left)} seconds", end=' ')
    
stream.stop_stream()
stream.close()
p.terminate()

print(f"\nRecorder>> Done recording...")

# --== Cria um novo objeto wave e guarda na diretoria indicada acima ==--

wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
wf.setnchannels(CHANNELS)
wf.setsampwidth(p.get_sample_size(FORMAT))
wf.setframerate(RATE)
wf.writeframes(b''.join(frames))
wf.close()

# --== Verifica se o ficheiro criado existe na diretoria ==--

if os.path.exists(WAVE_OUTPUT_FILENAME):
    print(f"\t   The wav file '{WAVE_OUTPUT_FILENAME}' has been created and saved in the directory.")
else:
    print("\t   There was a problem saving the WAV file.")
```

```python
# --== Imports ==--

import wave
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import librosa.display
import tensorflow as tf
from PIL import Image
import matplotlib.image as mpimg
%matplotlib inline 

L_max_w = 44100
ini_corta =int(L_max_w/10)
som = "0"
palavras = "zeros"
Lsubarray = 7000
detecao = 0.15

# --== Carrega o arquivo de audio para a variavel 'audio' ==--

with wave.open("audio//"+palavras+".wav", "rb") as audio:
    nchannels, sampwidth, framerate, nframes, comptype, compname = audio.getparams()
    samples = np.frombuffer(audio.readframes(nframes), dtype=np.int16)

# --== Mostra o gráfico do audio ==--

print("Audio Inteiro:")

plt.figure(figsize=(12, 4))
plt.plot(samples)
plt.title('Waveplot '+palavras)
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.show()

# --== Divide o audio em segmentos para analisar depois ==--

Lmax = int(len(samples)/Lsubarray) #transforma num int, o valor do tamanho do array que contem os frames do audio dividido por ???
samples_m = np.zeros(len(samples)) #Cria um array do tamanho da variavel samples só com 0's
samples = samples[ini_corta:]

for i in range(0, Lmax-1):
    start_n = i*Lsubarray
    end_n = start_n + Lsubarray
    std_val = np.std(samples[start_n:end_n]) # Calculo do desvio padrão
    samples_m[start_n:end_n] = std_val # Cria um subarray do desvio padrão calculado  
    
# --== Exibe o gráfico do áudio ==--

plt.figure(figsize=(12, 4))
plt.plot(samples_m)
plt.title('Waveplot envolvente: '+palavras)
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.show()    

# --== Define o threshold de silêncio ==--

threshold = detecao*np.average(samples_m)

print("--------------A dividir o audio por segmentos----------------------\n")

# --== Divide o áudio em segmentos onde o numero é dito ==--

segments = []
start = 0
for i in range(1, len(samples)):
    if abs(samples_m[i]) > threshold and abs(samples_m[i-1]) <= threshold:
        start = i
    elif abs(samples_m[i]) <= threshold and abs(samples_m[i-1]) > threshold:
        segments.append((start, i))

#estas linhas definem as caracteristicas da imagem do espectograma
hl = 8 #16 #512 # number of samples per time-step in spectrogram
hi = 256 #1024 #1024 #1024 #128 # Height of image
wi = 1024 #512 #4*384 # Width of image

# --== Exibe o gráfico de cada numero dito ==--

for i in range(0, len(segments)):
    print('Segmento de audio nº'+str(i))
    plt.figure(figsize=(12, 4))
    plt.plot(samples[segments[i][0]:segments[i][1]])
    plt.title('Waveplot data '+str(i))
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.show()
    dados_wave = samples[segments[i][0]:segments[i][1]]
    dif_len = L_max_w - len(dados_wave)
    if dif_len > 0:
        Tconcat = tf.concat([dados_wave, np.zeros(dif_len)], 0)
        dados_wave = tf.roll(Tconcat, int(dif_len/2), axis=0)
    else:
        dados_wave = dados_wave[:L_max_w]
    fig, ax = plt.subplots()

    y = tf.reshape(dados_wave, [L_max_w,]).numpy()
    y = y.astype(np.float64) 
    window = y[0:L_max_w]  # IGUAL A[0:wi*hl]
    S = librosa.feature.melspectrogram(y=window, sr=framerate, n_mels=hi, fmax=int(framerate/2), hop_length=hl)
    S_dB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=framerate, fmax=int(framerate/2), ax=ax)
    plt.savefig("audio//"+som+"_"+str(i)+"_ini.png")
    plt.show()

    # Itera sobre os segmentos e salva cada um como um arquivo de áudio separado
    with wave.open("audio//"+som+"_{}.wav".format(i), "wb") as output:
        output.setparams(audio.getparams())
        output.writeframes(samples[segments[i][0]:segments[i][1]].tobytes())
```

```python
# --== Imports ==-- 

%reload_ext tensorboard

import os, sys, math, datetime, pathlib2

import PIL
import PIL.Image

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras import callbacks
from keras.models import Sequential
from keras.layers import Conv2D, Dense, Flatten, Dropout, MaxPooling2D

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

image_height = int(432/2.5)
image_width = int(576/2.5)
n_channels = 3

max_poolSize = (2,2)
filter_size = 3
n_neuronios = 32
n_classes = 10
batch_size = 3+3

N_Neuronio1 = 4
N_Neuronio2 = 64+16

n_epochs = 200
# --== Carrega as imagens a usar pra CNN das pastas ==--

train_ds = tf.keras.utils.image_dataset_from_directory(
    directory="audio\\train",
    validation_split=0.02,
    subset="training",
    #shuffle=True,
    seed=123,
    image_size=(image_height, image_width),
    batch_size=batch_size)

val_ds = tf.keras.utils.image_dataset_from_directory(
    directory='audio\\validation',
    validation_split=0.99,
    subset="validation",
    #shuffle=True,
    seed=123,
    image_size=(image_height, image_width),
    batch_size=batch_size)

class_names = train_ds.class_names #No class_names fica um array com os valores de 0 a 9

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(image_height, image_width, n_channels)))
model.add(tf.keras.layers.Conv2D(n_neuronios, filter_size, strides=2, padding='same', activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.MaxPooling2D(pool_size=max_poolSize))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Conv2D(n_neuronios*2, filter_size, padding='same', activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.MaxPooling2D(pool_size=max_poolSize))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Conv2D(n_neuronios*4, filter_size, padding='same', activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.MaxPooling2D(pool_size=max_poolSize))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(n_neuronios*8, activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(n_classes, activation='softmax'))

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy'],
)

earlystopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    min_delta=0.96,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=0.96,
    restore_best_weights=True,
    start_from_epoch=130
) #https://keras.io/api/callbacks/early_stopping/

history = model.fit(train_ds, epochs=n_epochs, validation_data=val_ds, verbose=2, callbacks=[earlystopping])
#voltar a por o codigo das logs
```

```python

# Plot the loss curves for training and validation.
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values)+1)

plt.figure(figsize=(8,6))
plt.plot(epochs, loss_values, color ='blue', label='Training loss')
plt.plot(epochs, val_loss_values, color ='red', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot the accuracy curves for training and validation.
acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
epochs = range(1, len(acc_values)+1)

plt.figure(figsize=(8,6))
plt.plot(epochs, acc_values, color ='blue', label='Training accuracy')
plt.plot(epochs, val_acc_values, color ='red', label='Validation accuracy')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show() #graficos de https://www.kaggle.com/code/christianlillelund/classify-mnist-audio-using-spectrograms-keras-cnn

print("Training loss min value: {}\nValidation loss min value: {}".format(round(min(loss_values), 4), round(min(val_loss_values), 4)))
print("Training accuracy max value: {}\nValidation accuracy max value: {}".format(round(max(acc_values), 4), round(max(val_acc_values), 4)))
```
